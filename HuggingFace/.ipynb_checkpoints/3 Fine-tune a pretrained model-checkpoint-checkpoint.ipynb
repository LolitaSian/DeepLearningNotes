ä½¿ç”¨é¢„è®­ç»ƒæ¨¡å‹æœ‰è¯¸å¤šå¥½å¤„ï¼Œæ¯”å¦‚å‡å°‘è®¡ç®—å¼€é”€ï¼Œè®©ä½ ä¸ç”¨ä»å¤´è®­ç»ƒå°±å¯ä»¥æ–¹ä¾¿çš„ä½¿ç”¨SOTAæ¨¡å‹ã€‚ğŸ¤—Transformersæä¾›äº†ä¼—å¤šé¢„è®­ç»ƒæ¨¡å‹ï¼Œå½“ä½ ä½¿ç”¨é¢„è®­ç»ƒæ¨¡å‹çš„æ—¶å€™ï¼Œä½ å¯ä»¥åœ¨ç‰¹å®šä»»åŠ¡çš„æ•°æ®é›†ä¸Šå¯¹å…¶è¿›è¡Œå¾®è°ƒã€‚

æ¥ä¸‹æ¥çœ‹ä¸€ä¸‹
- å¾®è°ƒğŸ¤—Transformersçš„é¢„è®­ç»ƒæ¨¡å‹
- åœ¨åŸç”Ÿpytorchä¸­å¾®è°ƒæ¨¡å‹

å¾®è°ƒä¸€ä¸ªé¢„è®­ç»ƒæ¨¡å‹ä¹‹å‰éœ€è¦å‡†å¤‡å¥½é¢„è®­ç»ƒæ¨¡å‹ã€‚ä¸‹è¾¹ä½¿ç”¨ä¸€ä¸‹Yelp Reviewsæ•°æ®é›†ã€‚

```python
from datasets import load_dataset

dataset = load_dataset("yelp_review_full")
dataset["train"][0]  # datasetç»“æ„å¦‚ä¸‹ï¼Œæ‰“å°è®­ç»ƒé›†çš„ç¬¬ä¸€ç»„æ•°æ®

'''
DatasetDict({
    train: Dataset({
        features: ['label', 'text'],
        num_rows: 650000
    })
    test: Dataset({
        features: ['label', 'text'],
        num_rows: 50000
    })
})
'''
```

å‰è¾¹åªè¯´äº†æ€ä¹ˆç”¨tokenizerå¤„ç†å‡ ä¸ªå¥å­ã€‚è€Œå¯¹äºæ•°æ®é›†ğŸ¤—Datasetsæä¾›äº†`map`æ–¹æ³•å°†tokenizerå¤„ç†åº”ç”¨åˆ°æ•´ä¸ªæ•°æ®é›†ã€‚

```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("bert-base-cased")


def tokenize_function(example):
    return tokenizer(example["text"], padding="max_length", truncation=True)
# exampleå¯ä»¥æ¢æˆåˆ«çš„è¯

tokenized_datasets = dataset.map(tokenize_function, batched=True)
```

å¦‚æœä½ åªæƒ³ç”¨æ•°æ®é›†çš„ä¸€ä¸ªå­é›†æ¥å‡å°‘å¾®è°ƒæ—¶é—´ï¼Œå¯ä»¥è¿™æ ·ï¼š


```python
small_train_dataset = tokenized_datasets["train"].shuffle(seed=42).select(range(1000))
small_eval_dataset = tokenized_datasets["test"].shuffle(seed=42).select(range(1000))
```

# å¾®è°ƒ
ğŸ¤— Transformersæä¾›äº†[`Trainer`](https://huggingface.co/docs/transformers/v4.20.1/en/main_classes/trainer#transformers.Trainer)ç±»ç”¨äºä¼˜åŒ–è®­ç»ƒTransformeræ¨¡å‹ã€‚æ”¯æŒloggingã€gradient accumulationã€mixed precision.

```python
from transformers import AutoModelForSequenceClassification

# åŠ è½½é¢„è®­ç»ƒæ¨¡å‹ï¼Œé€‰æ‹©æƒ³è¦çš„æ ‡ç­¾æ•°é‡ï¼Œæ ¹æ®æˆ‘ä»¬ä½¿ç”¨çš„æ•°æ®é›†çš„ä»‹ç»å¯ä»¥çŸ¥é“Yelp Reviewçš„æ•°æ®é›†å¯é€‰æ ‡ç­¾ä¸º1~5ä¸ª
model = AutoModelForSequenceClassification.from_pretrained("bert-base-cased", num_labels=5)

'''
ä½ å¯èƒ½ä¼šçœ‹åˆ°è­¦å‘Šï¼Œè¯´ä¸€äº›é¢„è®­ç»ƒå‚æ•°æ²¡è¢«ç”¨åˆ°ï¼Œè¿˜æœ‰ä¸€äº›å‚æ•°è¢«éšæœºåˆå§‹åŒ–äº†ã€‚
è¿™æ˜¯å› ä¸ºåŸæ¥bertçš„è®­ç»ƒå¤´è¢«å¿½è§†æ‰äº†ï¼Œç”¨åˆ†ç±»å¤´å–ä»£äº†ï¼Œä¹‹åå°±æ˜¯è¦åœ¨åˆ†ç±»ä»»åŠ¡ä¸Šè¿›è¡ŒBEETæ¨¡å‹çš„å¾®è°ƒã€‚
'''
```

## Training hyperparameters

ğŸ¤—æä¾›äº†ä¸€ä¸ª[`TrainingArguments`ç±»](https://huggingface.co/docs/transformers/v4.20.1/en/main_classes/trainer#transformers.TrainingArguments)ï¼Œå…¶ä¸­åŒ…å«æ‰€æœ‰éœ€è¦è°ƒå‚çš„è¶…å‚æ•°ã€‚ ä¸‹è¾¹ä»£ç æ˜¯ç›´æ¥ä½¿ç”¨è¿™ä¸ªç±»çš„é»˜è®¤å‚æ•°ï¼Œä½ ä¹Ÿå¯ä»¥é€‰å‡ºè‡ªå·±éœ€è¦çš„è¿›è¡Œä¿®æ”¹ä»¥æ‰¾åˆ°æœ€ä½³å‚æ•°ã€‚

```python
from transformers import TrainingArguments

training_args = TrainingArguments(output_dir="test_trainer")
```

Traineræœ¬èº«æ˜¯ä¸å¸¦æœ‰è¯„ä»·æŒ‡æ ‡çš„ï¼Œåªæ˜¯ä¸€ä¸ªå¾ªç¯è®­ç»ƒã€‚ä½ éœ€è¦è‡ªå·±å¼•å…¥è¯„ä»·æŒ‡æ ‡ã€‚

ğŸ¤—æä¾›äº†ä¸€äº›å¸¸ç”¨çš„è¯„ä»·æŒ‡æ ‡ï¼Œç”¨[`load_metric`](https://huggingface.co/docs/datasets/metrics)å¯ä»¥ç›´æ¥åŠ è½½è¿›æ¥ã€‚
æ›´å¤šè¯„ä»·æŒ‡æ ‡å¯ä»¥çœ‹[`datasets.list_metrics`](https://huggingface.co/docs/datasets/v2.3.2/en/package_reference/loading_methods#datasets.list_metrics)ã€‚

è°ƒç”¨metricçš„computeæ–¹æ³•å³å¯å¯¹é¢„æµ‹ç»“æœè¿›è¡Œè¯„æµ‹ï¼Œä½†æ˜¯æ³¨æ„åœ¨è®¡ç®—ä¹‹å‰è¦å°†é¢„æµ‹ç»“æœè½¬åŒ–ä¸ºlogitsï¼Œå› ä¸ºğŸ¤—transformeré»˜è®¤è¿”å›å€¼éƒ½æ˜¯logistã€‚

æŒ‡å®š`evaluation_strategy`å‚æ•°å°±å¯ä»¥åœ¨æ¯ä¸ªepochç»“å°¾æ˜¾ç¤ºå½“å‰è®­ç»ƒç»“æœæŒ‡æ ‡ã€‚

```python
import numpy as np
from datasets import load_metric

metric = load_metric("accuracy")

def compute_metrics(eval_pred):
    logits, labels = eval_pred
    predictions = np.argmax(logits, axis=-1)
    return metric.compute(predictions=predictions, references=labels)

training_args = TrainingArguments(output_dir="test_trainer", evaluation_strategy="epoch")
```

æä¸€ä¸ªTrainerå¯¹è±¡æŠŠåˆšæ‰çš„æ¨¡å‹ã€è®­ç»ƒå‚æ•°ã€è®­ç»ƒé›†æµ‹è¯•é›†æ•°æ®ã€è¯„ä»·æŒ‡æ ‡ç»„åˆåˆ°ä¸€èµ·ã€‚å°±å¯ä»¥å¾®è°ƒäº†ã€‚

```python
from transformers import Trainer

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=small_train_dataset,
    eval_dataset=small_eval_dataset,
    compute_metrics=compute_metrics,
)

trainer.train()
```

# ç”¨åŸç”Ÿpytorchå¾®è°ƒ

`Trainer`åªæ˜¯æä¾›ä¸€ä¸ªè¿­ä»£æ–¹æ³•ï¼Œè®©ä½ å¯ä»¥ä¸€è¡Œä»£ç è®­ç»ƒæ¨¡å‹ã€‚å½“ç„¶å¦‚æœä½ æƒ³è‡ªå·±å†™å¾ªç¯ä¹Ÿå®Œå…¨å¯ä»¥ã€‚ä½ å¯ä»¥æ‰‹å†™ä»£ç è¿›è¡Œå¾®è°ƒã€‚

é¦–å…ˆæ¸…ç©ºä¸€ä¸‹ä¸Šè¾¹åˆšæ‰æ‰§è¡Œçš„ä¸œè¥¿ã€‚


```python
del model
del trainer
torch.cuda.empty_cache()
```

## æ•°æ®é¢„å¤„ç†
æ‰‹å†™é¢„å¤„ç†çš„`tokenizer_dataset`ï¼Œç„¶åæŠ½å–æ•°æ®é›†çš„å­é›†è¿›è¡Œè®­ç»ƒã€‚

```python
# 1.ç§»é™¤textåˆ—ï¼Œå› ä¸ºæ¨¡å‹ä¸èƒ½æ¥å—åŸå§‹æ–‡æœ¬ä½œä¸ºè¾“å…¥
tokenized_datasets = tokenized_datasets.remove_columns(["text"])
# 2.é‡å‘½å`label`ä¸º`labels`ï¼Œå› ä¸ºæ¨¡å‹å¸Œæœ›å‚æ•°æ˜¯`labels`
tokenized_datasets = tokenized_datasets.rename_column("label", "labels")
# 3.å°†æ ¼å¼è½¬åŒ–ä¸ºpytorchçš„å¼ é‡
tokenized_datasets.set_format("torch")

# æŠ½å­é›†
small_train_dataset = tokenized_datasets["train"].shuffle(seed=42).select(range(1000))
small_eval_dataset = tokenized_datasets["test"].shuffle(seed=42).select(range(1000))
```

## DataLoader
ä¸ºä½ çš„è®­ç»ƒé›†å’Œæµ‹è¯•é›†åˆ›å»ºä¸€ä¸ª`dataloader`ï¼Œæ–¹ä¾¿åœ°å–åˆ°æ¯ä¸ªbatchçš„æ•°æ®ã€‚

```python
from torch.utils.data import DataLoader

train_dataloader = DataLoader(small_train_dataset, shuffle=True, batch_size=8)
eval_dataloader = DataLoader(small_eval_dataset, batch_size=8)
```

æ ¹æ®ä½ éœ€è¦çš„æ ‡ç­¾æ•°é‡åŠ è½½æ¨¡å‹ã€‚

```python
from transformers import AutoModelForSequenceClassification

model = AutoModelForSequenceClassification.from_pretrained("bert-base-cased", num_labels=5)
```

## Optimizer and learning rate scheduler
åˆ›å»ºä¼˜åŒ–å™¨å’Œå­¦ä¹ ç‡è°ƒèŠ‚å™¨æ¥å¾®è°ƒæ¨¡å‹ï¼Œè¿™é‡Œæˆ‘ä»¬ä½¿ç”¨AdamWã€‚

ä»Traineråˆ›å»ºé»˜è®¤çš„å­¦ä¹ é€Ÿç‡è°ƒåº¦å™¨ã€‚

æœ€åæŠŠä»£ç æ‰”åˆ°GPUä¸Šã€‚å¦‚æœæœ‰GPUçš„è¯ã€‚

```python
from torch.optim import AdamW
optimizer = AdamW(model.parameters(), lr=5e-5)

from transformers import get_scheduler
num_epochs = 3
num_training_steps = num_epochs * len(train_dataloader)
lr_scheduler = get_scheduler(
    name="linear", optimizer=optimizer, num_warmup_steps=0, num_training_steps=num_training_steps
)

import torch
device = torch.device("cuda") if torch.cuda.is_available() else torch.device("cpu")
model.to(device)
```

Great, now you are ready to train! ğŸ¥³

## è®­ç»ƒå¾ªç¯

ä¸ºäº†èƒ½ç›‘è§†ä½ çš„è®­ç»ƒè¿›åº¦ï¼Œä½¿ç”¨`tqdm`åº“ç»™è®­ç»ƒåŠ ä¸€ä¸ªè¿›åº¦æ¡ã€‚

```python
from tqdm.auto import tqdm

progress_bar = tqdm(range(num_training_steps))

model.train()

for epoch in range(num_epochs):
    for batch in train_dataloader:
        batch = {k: v.to(device) for k, v in batch.items()}
        outputs = model(**batch)
        loss = outputs.loss
        loss.backward()

        optimizer.step()
        lr_scheduler.step()
        optimizer.zero_grad()
        progress_bar.update(1)
```

## è¯„ä»·æŒ‡æ ‡

å’Œä¹‹å‰çš„ä¸€æ ·ï¼Œä½ éœ€è¦ç»™ä½ è‡ªå·±å†™çš„å¾ªç¯åŠ ä¸€ä¸ªè®­ç»ƒæŒ‡æ ‡ã€‚ä½†æ˜¯ï¼Œè¿™æ¬¡ä¸æ˜¯åœ¨æ¯ä¸ªepochç»“æŸæ—¶è®¡ç®—å’ŒæŠ¥å‘Šåº¦é‡ï¼Œè€Œæ˜¯ä½¿ç”¨add_batchç´¯è®¡æ‰€æœ‰æ‰¹ï¼Œå¹¶åœ¨æœ€åè®¡ç®—åº¦é‡ã€‚



```python
metric = load_metric("accuracy")
model.eval()
for batch in eval_dataloader:
    batch = {k: v.to(device) for k, v in batch.items()}
    with torch.no_grad():
        outputs = model(**batch)

    logits = outputs.logits
    predictions = torch.argmax(logits, dim=-1)
    metric.add_batch(predictions=predictions, references=batch["labels"])

metric.compute()
```
