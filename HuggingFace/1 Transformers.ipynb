## pipeline

`pipeline()`可以让你方便地使用[model Hub](https://huggingface.co/models)

`pipline()`更完整的功能可以看[pipeline文档](https://huggingface.co/docs/transformers/v4.20.1/en/main_classes/pipelines#transformers.pipeline

用pipeline做一生成任务：

```python
from transformers import pipeline
```

```python
generator = pipeline(task="text-generation")
generator(
    "Hello, Sian"
)  # doctest: +SKIP

# 如果要传入多个输入，使用列表存储。
```

还有一些参数可以选择，比如输出两个结果：

```python
generator(
    "Hello, Sian",
    num_return_sequences=2,
)  # doctest: +SKIP
```

前边提示如果你不指定模型会使用默认的xxx模型。所以这意思是pipeline可以指定模型。

所以你可以从model hub中找到你需要的模型之后加载模型。

最简单的指定方法：

```python
from transformers import pipeline
generator = pipeline('text-generation', model='distilgpt2')
```

使用对应的`AutoModelFor`和`AutoTokenizer`类去加载。

文档见[Auto Classes](https://huggingface.co/docs/transformers/v4.20.1/en/model_doc/auto#transformers.AutoModelForCausalLM)

```python
from transformers import AutoTokenizer, AutoModelForCausalLM

tokenizer = AutoTokenizer.from_pretrained("distilgpt2")
model = AutoModelForCausalLM.from_pretrained("distilgpt2")
generator = pipeline(task="text-generation", model=model, tokenizer=tokenizer)

generator("Hello, Sian")
```

或者使用对应的加载方式，这个在对应模型介绍页面会详细写如何调用。

```python
from transformers import GPT2Tokenizer, GPT2Model

tokenizer = GPT2Tokenizer.from_pretrained('distilgpt2')
model = GPT2Model.from_pretrained('distilgpt2')

text = "Replace me by any text you'd like."

encoded_input = tokenizer(text, return_tensors='pt')

output = model(**encoded_input)
```
